## Knn:--

```python
from sklearn.neighbors import KNeighborsClassifier

# In the following line, "knn" is instantiated as an "object" of KNeighborsClassifier "class". 

k = 1
knn = KNeighborsClassifier(n_neighbors=k)

# Training ONLY on the training set:

knn.fit(X_train, y_train)
```

```python
# Testing on the testing set:

y_predict_knn = knn.predict(X_test)

print(y_predict_knn)
```

```python
from sklearn.metrics import r2_score
knn_accuracy = r2_score(y_test,y_predict_knn)
print(knn_accuracy)
```

```python
for i in range(1,16):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    knn_pred = knn.predict(X_test)
    print("Accuracy for {} neighbours".format(i), r2_score(y_test, knn_pred))
```

## Decision Tree:--

```python
# The following line will import DecisionTreeClassifier "Class"
# DecisionTreeClassifier is name of a "sklearn class" to perform "Decision Tree Classification" 

from sklearn.tree import DecisionTreeClassifier

# "my_decisiontree" is instantiated as an "object" of DecisionTreeClassifier "class". 

my_decisiontree = DecisionTreeClassifier()

# We can use the method "fit" of the objects "my_decisiontree" along with training dataset and labels to train the model.

my_decisiontree.fit(X_train, y_train)
```

```python
# Testing on the testing set:

y_predict_decisiontree = my_decisiontree.predict(X_test)

print(y_predict_decisiontree)
```

```python
from sklearn.metrics import r2_score
decisiontree_accuracy = r2_score(y_test,y_predict_decisiontree)
decisiontree_accuracy
```

```python
from sklearn import metrics

# Calculating "Mean Square Error" (MSE):
mse = metrics.mean_squared_error(y_test, y_predict_decisiontree)

# Using numpy sqrt function to take the square root and calculate "Root Mean Square Error" (RMSE)
rmse = np.sqrt(mse)

print(rmse)
```

### Cross Validation:--

```python
from sklearn.cross_validation import cross_val_score

my_decisiontree = DecisionTreeClassifier()

r2_score_list = cross_val_score(my_decisiontree, X, y, cv=10,scoring='r2')

r2_score_list
```

```python
# use average of accuracy values as final result
accuracy_cv = r2_score_list.mean()

print(accuracy_cv)
```

## Random Forest:--

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
my_RandomForest = RandomForestRegressor(n_estimators = 19, bootstrap = True, random_state=2)
my_RandomForest.fit(X_train, y_train)
```

```python
# Testing on the testing set:

y_predict_randomforest = my_RandomForest.predict(X_test)

print(y_predict_randomforest)
```

```python
from sklearn.metrics import r2_score
RandomForest_accuracy = r2_score(y_test,y_predict_randomforest)
RandomForest_accuracy
```

```python
from sklearn.model_selection import model_selection

seed = 7
num_trees = 120
max_features = 10
kfold2 = model_selection.KFold(n_splits=10, random_state=seed)
model2 = RandomForestRegressor(n_estimators=num_trees, max_features=max_features)
results2 = model_selection.cross_val_score(model2, X, y, cv=kfold2)
print(results2.mean())
```
